#INPUTS
Below is a set of input variables you need to enter values for. Example inputs are below. Note that the password you have to enter is the password for the github repo will eventually push to. Also, googledriveFolderPathWhereRepoWillBeCloned variable is the folder where the repo will be cloned in your github.

```
#INPUTS

#your github email and username
userEmail = 'sanmesh1@gmail.com'
userName = 'sanmesh1'

#gitHttpsWebUrl doesn't have to be a github repo you own
gitHttpsWebUrl = 'https://github.com/sanmesh1/TextSimilarityUsingWord2Vec.git' 
repoName = 'TextSimilarityUsingWord2Vec'

#enter your github password
from getpass import getpass
password = getpass('Password:')
import urllib.parse
password = urllib.parse.quote(password)

#googledriveFolderUrlToRepo should start with the path 
#'/content/gdrive/My Drive/' and go to the folder where you will 
#save the git repo:
googledriveFolderPathWhereRepoWillBeCloned = '/content/gdrive/My Drive/TestingGitCloneIntoGoogleDrive'


```

Run the below to do some prepreocessing

```
#Preprocessing
repoPath = googledriveFolderPathWhereRepoWillBeCloned + '/' + repoName
cloneUrl = 'https://'+userName+':'+ password +'@' + gitHttpsWebUrl[8:]
```

Run the below to mount your google drive onto the colab server so that your google drive folders may be accessible

```
from google.colab import drive
drive.mount('/content/gdrive')
```

Run the below to configure your github and to clone the repo into you google drive folder

```
!git config --global user.email $userEmail
!git config --global user.name $userName

#%cd gdrive/My Drive/project_folder
%cd $googledriveFolderPathWhereRepoWillBeCloned
#! git clone https://github.com/sanmesh1/TextSimilarityUsingWord2Vec.git
! git clone $cloneUrl
```

Run the below cell to enter into your repo folder and git pull to see if there are any updates

```
#%cd TextSimilarityUsingWord2Vec
%cd $repoPath
! git pull
```
